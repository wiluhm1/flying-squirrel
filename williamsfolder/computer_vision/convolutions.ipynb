```python
def make_windows(sequence, windowsize):
    positions = len(sequence) - windowsize + 1
    windows = []
    for i in range(positions):
        windows.append(sequence[i:i+windowsize])
    return windows

def print_padded_seq(seq):
    print("[", ",".join(["{:4d}".format(i) for i in seq]), ']')

def print_sliding_windows(seq, windowsize=3):
    windows = make_windows(seq, windowsize)
    for window in windows:
        print(",".join(["{:4d}".format(i) for i in window]))
   
```

```python
series = [0, 1, 0, 2, 1, 0, 1, 101, 100, 98, 102, 101]
windowsize = 2

print_padded_seq(series)

print_sliding_windows(series, windowsize=windowsize)
#check(1)
```

```python
convolutions = []
kernel = np.array([-1,1])
for w in make_windows(series, windowsize=2):
    w = np.array(w)
    convolved = np.dot(w,kernel)
    convolutions.append(convolved)
    
plt.plot(convolutions);
#check(2)
```

**Question:** Why does the graph move up at the end?

```python
#A convolution is essentially a third function that graphs the overlap of two functions to the value of the horizontal shift of the functions
#that corresponding to that overlap. There is an initial function that is left unchanged and a second function that is reflected across the
#y-axis. np.convolve works by taking two parameters which are both one dimensional arrays. These arrays are meant to represent functions that
#have x-coordinates of the position of each element and y-coordinates with the value of each element. np.convolve first reflects the second
#array across the y-axis and then calculates the overlap of that array with the first array by summing the product of the corresponding
#coordinates of the arrays. This value is the value of the convolution at that position on the graph.

#The sharp climb at the end can be explained by what happens when the array [-1, 1] is convolved with ys. At that point, the dataset ys,
#after being reflected across the y-axis, is shifted all the way to the right. 101 is the rightmost point in the initial ys dataset, but
#after being reflected across the y-axis and shifted to the right, 101 is the only point that can possibly overlap with [-1, 1]. -1 is
#multiplied by 0 because there is nothing from the ys dataset to overlap with while 1 is multipled by 101 because they occupy the same
#x-value. Therefore, the overlap should be 101 at that point, which is visible in the graph.
```
